# DDPM: Denoising Diffusion Probabilistic Models

Denoising Diffuison Probabilistic Models, Ho et al., 2020. [[arXiv:2006.11239](https://arxiv.org/abs/2006.11239)]

## 1. Prepare Dataset

Define the gaussian mixture distribution as the target distribution.

```python
import matplotlib.pyplot as plt
import numpy as np
import torch
import torch.distributions as D

mixture = D.MixtureSameFamily(
    D.Categorical(torch.tensor([0.3, 0.7])),
    D.Normal(torch.tensor([-0.5, 1.]), torch.tensor([0.1, 0.2])),
)
# target distribution
X = mixture.sample((100000, 1))

_ = plt.hist(X, bins=np.linspace(-5, 5, 200), label="data")
```

## 2. Define Model

Define MLP backbone and DDPM framework.

```python
import torch
import torch.nn as nn

from flowmodels import DDPM, DDIMScheduler


class Backbone(nn.Module):
    def __init__(self, dim: int = 1):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(dim + 1, 64),
            nn.ReLU(),
            nn.Linear(64, 64),
            nn.ReLU(),
            nn.Linear(64, dim)
        )

    def forward(self, x_t: torch.Tensor, t: torch.Tensor) -> torch.Tensor:
        return self.net.forward(torch.cat([x_t, t[:, None].to(x_t)], dim=-1))

scheduler = DDIMScheduler()
model = DDPM(Backbone(), scheduler)
model
```

```python
# intermediate plots
t = scheduler.T - 1
alpha_bar = 1 - scheduler.var()

eps = torch.randn_like(X)
noised = model.noise(X, torch.full((len(X),), t, dtype=torch.long), eps=eps)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)

# 1D hists
plt.hist(X, bins=np.linspace(-5, 5, 200), label="data")
plt.hist(noised, bins=np.linspace(-5, 5, 200), label="noised", alpha=0.7)
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(alpha_bar, label="\\bar\\alpha")
plt.legend()
```

## 3. Train DDPM

Train the diffusion model with the prepared dataset.

```python
import torch
from tqdm.auto import tqdm

# training configurations
LR = 0.001
TRAIN_STEPS = 1000
BATCH_SIZE = 2048

# train
model.train()
optim = torch.optim.Adam(model.parameters(), LR)

losses = []
with tqdm(range(TRAIN_STEPS)) as pbar:
    for i in pbar:
        loss = model.loss(sample=X[torch.randint(0, len(X), (BATCH_SIZE,))])
        # update
        optim.zero_grad()
        loss.backward()
        optim.step()

        # log
        loss = loss.detach().item()
        losses.append(loss)
        pbar.set_postfix_str(f"loss: {loss:.2f}")

plt.plot(losses)
```

## 4. Test the model

Sample from the model, inspect the trajectory.

```python
n = 10000
sample = torch.randn(n, X.shape[-1])
x_t, x_ts = model.sample(sample, verbose=lambda x: tqdm(x, leave=False))

plt.hist(sample, bins=np.linspace(-5, 5, 200), label="prior")
plt.hist(X[:n], bins=np.linspace(-5, 5, 200), label="gt", alpha=0.7)
plt.hist(x_t.view(-1), bins=np.linspace(-5, 5, 200), label="ddpm", alpha=0.5)
plt.legend()
_xticks, _ = plt.xticks()
```

```python
# Sampling trajectory
n = 200
x_t, x_ts = model.sample(torch.linspace(-3, 3, n)[:, None])
for i in range(10):
    plt.plot(torch.tensor([_x_t[i * 20].item() for _x_t in x_ts]), range(len(x_ts)))
    plt.xticks(_xticks)
```

### DDIM: Denoising Diffusion Implicit Models

Denoising Diffusion Implicit Models, Song et al., 2020. [(https://arxiv.org/abs/2010.02502)]

```python
# DDIM: Denoising Diffusion Implicit Models, Song et al., 2020. [arXiv:2010.02502]
from flowmodels import DDIMSampler

ddim = DDIMSampler(model.scheduler)

n = 10000
prior = torch.randn(n, X.shape[-1])
x_t, x_ts = ddim.sample(model, prior)

plt.figure(figsize=(5, 10))
plt.subplot(2, 1, 1)
plt.hist(prior, bins=np.linspace(-5, 5, 200), label="prior")
plt.hist(X[:n], bins=np.linspace(-5, 5, 200), label="gt", alpha=0.7)
plt.hist(x_t.view(-1), bins=np.linspace(-5, 5, 200), label="ddim", alpha=0.5)
plt.legend()
xticks, _ = plt.xticks()

plt.subplot(2, 1, 2)
for i in range(10):
    plt.plot(torch.tensor([_x_t[i * 20].item() for _x_t in x_ts]), range(len(x_ts)),)
    plt.xticks(xticks)
```

### PF-ODE: Probability Flow ODE
Score-Based Generative Modeling through Stochastic Differential Equations, Song et al., 2020. [[arXiv:2011.13456](https://arxiv.org/abs/2011.13456)]

```python
# PF ODE: Score-Based Generative Modeling through Stochastic Differential Equations, Song et al., 2020. [arXiv:2011.13456]
from flowmodels import ProbabilityFlowODESampler


sampler = ProbabilityFlowODESampler(model.scheduler)

n = 10000
prior = torch.randn(n, X.shape[-1])
x_t, x_ts = sampler.sample(model, prior)

plt.figure(figsize=(5, 10))
plt.subplot(2, 1, 1)
plt.hist(prior, bins=np.linspace(-5, 5, 200), label="prior")
plt.hist(X[:n], bins=np.linspace(-5, 5, 200), label="gt", alpha=0.7)
plt.hist(x_t.view(-1), bins=np.linspace(-5, 5, 200), label="PF-ODE", alpha=0.5)
plt.legend()
xticks, _ = plt.xticks()

plt.subplot(2, 1, 2)
for i in range(10):
    plt.plot(torch.tensor([_x_t[i * 20].item() for _x_t in x_ts]), range(len(x_ts)),)
    plt.xticks(xticks)
```
