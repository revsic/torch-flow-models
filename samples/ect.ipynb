# Easy Consistency Training

Consistency Models Made Easy, Geng et al., ICLR 2025. [arXiv:2406.14548](https://arxiv.org/abs/2406.14548)

## 1. Prepare Dataset

Define the gaussian mixture distribution as the target distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

from utils import gmm

X, Y = gmm(100000, weights=[0.3, 0.7], mean=[-0.5, 1.0], std=[0.1, 0.2])

_ = plt.hist(X, bins=np.linspace(-5, 5, 200), label="data")
```

## 2. Define Model

Define MLP backbone and ECT framework.

```python
from flowmodels.ect import EasyConsistencyTraining, ECTScheduler
from utils import Backbone


TRAIN_STEPS = 1000
COND = False
aux = 1 if not COND else 2
model = EasyConsistencyTraining(
    Backbone(aux=aux), ECTScheduler(total_training_steps=TRAIN_STEPS)
)
model
```

## 3. Training Consistency Models

Train the consistency model with the prepared dataset.

```python
import torch
from tqdm.auto import tqdm

from utils import vis

# training configurations
LR = 0.001
BATCH_SIZE = 2048
MU = 0.0

# train
model.train()
optim = torch.optim.Adam(model.parameters(), LR)

losses = []
with tqdm(range(TRAIN_STEPS)) as pbar:
    for i in pbar:
        indices = torch.randint(0, len(X), (BATCH_SIZE,))
        loss = model.loss(
            sample=X[indices],
            label=Y[indices] if COND else None,
        )
        # update
        optim.zero_grad()
        loss.backward()
        optim.step()

        # log
        loss = loss.detach().item()
        losses.append(loss)
        pbar.set_postfix_str(f"loss: {loss:.2f}")

plt.plot(losses)
vis(model, conditional=COND, steps=[None])
```
