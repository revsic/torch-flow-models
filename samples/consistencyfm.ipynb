# Consistency Flow Matching

Defining Straight Flows with Velocity Consistency, Yang et al., 2024. [[arXiv:2407.02398](https://arxiv.org/abs/2407.02398)]

## 1. Prepare Dataset

Define the gaussian mixture distribution as the target distribution.

```python
import matplotlib.pyplot as plt
import numpy as np
import torch
import torch.distributions as D

mixture = D.MixtureSameFamily(
    D.Categorical(torch.tensor([0.3, 0.7])),
    D.Normal(torch.tensor([-0.5, 1.0]), torch.tensor([0.1, 0.2])),
)
# target distribution
X = mixture.sample((100000, 1))

_ = plt.hist(X, bins=np.linspace(-5, 5, 200), label="data")
```

## 2. Define Model

Define MLP backbone and consistency FM framework.

```python
import torch
import torch.nn as nn

from flowmodels import ConsistencyFlowMatching


class Backbone(nn.Module):
    def __init__(self, dim: int = 1, layers: int = 3):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(dim + 1, 64),
            *[
                submodule
                for _ in range(layers)
                for submodule in (nn.ReLU(), nn.Linear(64, 64))
            ],
            nn.ReLU(),
            nn.Linear(64, dim),
        )

    def forward(self, x_t: torch.Tensor, t: torch.Tensor) -> torch.Tensor:
        return self.net.forward(torch.cat([x_t, t[:, None].to(x_t)], dim=-1))


model = ConsistencyFlowMatching(Backbone())
model
```

## 3. Train Consistency FM

Train the ODE model with the prepared dataset.

```python
import torch
from tqdm.auto import tqdm

from flowmodels.consistencyfm import EMASupports

# training configurations
LR = 0.001
TRAIN_STEPS = 1000
BATCH_SIZE = 2048
MU = 0.0

# train
model.train()
optim = torch.optim.Adam(model.parameters(), LR)
# EMA Supports
ema = EMASupports(model)

losses = []
with tqdm(range(TRAIN_STEPS)) as pbar:
    for i in pbar:
        loss = model.loss(
            sample=X[torch.randint(0, len(X), (BATCH_SIZE,))],
            ema=ema,
        )
        # update
        optim.zero_grad()
        loss.backward()
        optim.step()

        # ema updat
        ema.update(model, mu=MU)

        # log
        loss = loss.detach().item()
        losses.append(loss)
        pbar.set_postfix_str(f"loss: {loss:.2f}")

plt.plot(losses)

# plot histogram
n = 10000
sample = torch.randn(n, X.shape[-1])
x_t, _ = model.sample(sample, verbose=lambda x: tqdm(x, leave=False), steps=100)
x_2, _ = model.sample(sample, verbose=lambda x: tqdm(x, leave=False), steps=2)

plt.figure()
plt.hist(sample, bins=np.linspace(-5, 5, 200), label="prior")
plt.hist(X[:n], bins=np.linspace(-5, 5, 200), label="gt", alpha=0.7)
plt.hist(x_t.view(-1), bins=np.linspace(-5, 5, 200), label="cfm", alpha=0.5)
plt.hist(x_2.view(-1), bins=np.linspace(-5, 5, 200), label="cfm-2", alpha=0.5)
plt.legend()
_xticks, _ = plt.xticks()

# plot trajectory
n = 200
_, x_ts = model.sample(torch.linspace(-3, 3, n)[:, None], steps=100)
_, x_2s = model.sample(torch.linspace(-3, 3, n)[:, None], steps=2)

plt.figure()
for i in range(10):
    plt.plot(
        torch.tensor([_x_t[i * 20].item() for _x_t in x_ts]),
        np.linspace(0, 1, len(x_ts)),
        "b",
        **({} if i > 0 else {"label": "cfm"}),  # labeling on first plot only
    )
    plt.plot(
        torch.tensor([_x_t[i * 20].item() for _x_t in x_2s]),
        np.linspace(0, 1, len(x_2s)),
        "r",
        alpha=0.7,
        **({} if i > 0 else {"label": "cfm-2"}),  # labeling on first plot only
    )
    plt.xticks(_xticks)
plt.legend()
```
