# Consistency Flow Matching

Defining Straight Flows with Velocity Consistency, Yang et al., 2024. [[arXiv:2407.02398](https://arxiv.org/abs/2407.02398)]

## 1. Prepare Dataset

Define the gaussian mixture distribution as the target distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

from utils import gmm

X, Y = gmm(100000, weights=[0.3, 0.7], mean=[-0.5, 1.0], std=[0.1, 0.2])

_ = plt.hist(X, bins=np.linspace(-5, 5, 200), label="data")
```

## 2. Define Model

Define MLP backbone and consistency FM framework.

```python
from flowmodels import ConsistencyFlowMatching
from utils import Backbone

COND = False
aux = 1 if not COND else 2
model = ConsistencyFlowMatching(Backbone(aux=aux))
model
```

## 3. Train Consistency FM

Train the ODE model with the prepared dataset.

```python
import torch
from tqdm.auto import tqdm

from flowmodels.utils import EMASupports
from utils import vis

# training configurations
LR = 0.001
TRAIN_STEPS = 1000
BATCH_SIZE = 2048
MU = 0.0

# train
model.train()
optim = torch.optim.Adam(model.parameters(), LR)
# EMA Supports
ema = EMASupports(model)

losses = []
with tqdm(range(TRAIN_STEPS)) as pbar:
    for i in pbar:
        indices = torch.randint(0, len(X), (BATCH_SIZE,))
        loss = model.loss(
            sample=X[indices],
            label=Y[indices] if COND else None,
            ema=ema,
        )
        # update
        optim.zero_grad()
        loss.backward()
        optim.step()

        # ema updat
        ema.update(model, mu=MU)

        # log
        loss = loss.detach().item()
        losses.append(loss)
        pbar.set_postfix_str(f"loss: {loss:.2f}")

plt.plot(losses)
vis(model, conditional=COND, steps=[100, 2])
```
